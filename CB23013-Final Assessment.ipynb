```python
# SVM
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
# Define features and target
X = df.drop(columns=['student_id', 'grade', 'burnout_score', 'high_burnout'])
y = df['high_burnout']

X = pd.get_dummies(X, drop_first=True)
```

```python
# Split train-test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create SVM model
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train_scaled, y_train)
```

```python
# Prediction
y_pred = svm_model.predict(X_test_scaled)
print("=== SVM Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
# Model accuracy
y_pred = svm_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Visualize confusion matrix
# Calculate the raw numbers for the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix visually using matplotlib
plt.figure(figsize=(6, 5))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Burnout', 'High Burnout'])
disp.plot(cmap='Blues', values_format='d')
plt.title('SVM Model Confusion Matrix')
plt.show()
```
